```python
# load the collected datapoints json files for each pose
df_downwarddog = pd.read_json('yoga-dataset/collected-datapoints/downwarddog_data.json')
df_mountain = pd.read_json('yoga-dataset/collected-datapoints/mountain_data.json')
df_tree = pd.read_json('yoga-dataset/collected-datapoints/tree_data.json')
# df_warrior1 = pd.read_json('yoga-dataset/collected-datapoints/warrior1_data.json')
df_warrior2 = pd.read_json('yoga-dataset/collected-datapoints/warrior2_data.json')

# loading additional collected data
df_downwarddog_2 = pd.read_json('yoga-dataset/collected-datapoints/downwarddog-2_data.json')
df_warrior2_2 = pd.read_json('yoga-dataset/collected-datapoints/warrior2-2_data.json')
df_tree_2 = pd.read_json('yoga-dataset/collected-datapoints/tree-2_data.json')
df_goddess = pd.read_json('yoga-dataset/collected-datapoints/goddess_data.json')
```

```python
# transpose the dataframes, then concat and then change the labels on the rows
dfs = [df_downwarddog, df_downwarddog_2, df_mountain, df_tree, df_tree_2, df_goddess, df_warrior2, df_warrior2_2]
idx = 0;
for i in range(len(dfs)):
    dfs[i] = dfs[i].transpose()
    idx += len(dfs[i])
```

```python
# of data
idx
```

```python
rows_list = []
for i in range(0, idx):
    rows_list.append(i)
```

```python
rows_list
```

```python
yoga_df = pd.concat(dfs)
```

```python
class_names = ['downward dog', 'mountain', 'tree', 'goddess', 'warrior2']
class_names
```

```python
labels = []
for i in range(0,len(dfs[0])+len(dfs[1])):
    labels.append(0)
for i in range(0,len(dfs[2])):
    labels.append(1)
for i in range(0,len(dfs[3])+len(dfs[4])):
    labels.append(2)
for i in range(0,len(dfs[5])):
    labels.append(3)
for i in range(0,len(dfs[6])+len(dfs[7])):
    labels.append(4)
```
